## a)
Build a classifier using a full connected three layer neural network with logistic activation functions. Your network should

• take a vector $ x ∈ R^{10}$ as input (nine features plus a constant offset)

• have a single, fully connected hidden layer with 32 neurons

• output a scalar $\hat{y}$.

Note that since the logistic activation function is always positive, your decision should be as follows: yb > 0.5 corresponds to a ‘happy’ face, while yb ≤ 0.5 is not happy.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
dataset = loadmat('face_emotion_data.mat')

X, y = dataset['X'], dataset['y']
n, p = np.shape(X)

y[y==-1] = 0  # use 0/1 for labels instead of -1/+1
X = np.hstack((np.ones((n,1)), X))  # append a column of ones
```

```python
def logsig(_x):
    return 1/(1+np.exp(-_x))



X_train = np.hstack((np.ones((n,1)), X))
Y_labels = y
## initial weights
n,p = np.shape(X_train)
M=32
L=200
alpha = .05

def train(training_index):
    
    training_error = []
    V = np.random.randn(M+1, 1); 
    W = np.random.randn(p, M);


    for epoch in range(L):
        ind = np.random.permutation(n)
        for i in ind:
            # Forward-propagate
            H = logsig(np.hstack((np.ones((1,1)), X_train[[i],:]@W)))
            Yhat = logsig(H@V)
             # Backpropagate
            delta = (Yhat-y[[i],:])*Yhat*(1-Yhat)
            Vnew = V-alpha*H.T@delta
            gamma = delta@V[1:,:].T*H[:,1:]*(1-H[:,1:])
            Wnew = W - alpha*X_train[[i],:].T@gamma
            V = Vnew
            W = Wnew

        ## Final predicted labels (on training data)
        H = logsig(np.hstack((np.ones((n,1)), X_train@W)))
        Yhat = logsig(H@V)
        training_error.append(np.mean(abs(np.round(Yhat[:,0])-y[:,0])))
    return(W,V, training_error)




def testing_data(testing_index,W,V):
    X_test = X[testing_index]
    y_test = y[testing_index]
    nt,_ = np.shape(X_test)
    H = logsig(np.hstack((np.ones((n,1)), X_train@W)))
    Yhat = logsig(H@V)
    return np.mean(abs(np.round(Yhat[:,0])-y[:,0]))

```
